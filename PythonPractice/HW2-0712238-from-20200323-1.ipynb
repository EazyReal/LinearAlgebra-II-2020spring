## 七、選取特定的DataFrame

- DataFrame.loc 
- DataFrame.iloc

The [**`loc`**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html) method is used to select rows and columns by **label**. You can pass it:

- A single label
- A list of labels
- A slice of labels
- A boolean Series
- A colon (which indicates "all labels")

The [**`iloc`**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html) method is used to select rows and columns by **integer position**. You can pass it:

- A single integer position
- A list of integer positions
- A slice of integer positions
- A colon (which indicates "all integer positions")

```python
# rows 10 through 20(inclusive) , columns ['Survived','Age','Fare']

train.loc[10:20, ['Survived','Age','Fare']]
```

```python
# all rows, columns ['Survived','Age','Fare']

train.loc[:, ['Survived','Age','Fare']]
```

```python
# rows 10 through 13(inclusive) , columns 'Survived through Sex(inclusive)

train.loc[10:13,'Survived':'Sex']
```

```python
train.loc[train.Survived == 1,'Survived':'Fare'].head()
```

```python
# rows in positions 10 through 20 (exclusive), columns in positions 1、5 and 9

train.iloc[10:20, [1,5,9]]
```

```python
train.head()
```

```python
# rows in positions 0 through 2 (exclusive), all columns
train.iloc[10:20, 0:3]
```

```python
train.iloc[0:2, :]
```

```python
train.iloc[0:2]
```

```python
# accomplish the same thing - but using 'iloc' is preferred since it's more explicit

train[0:2]
```
